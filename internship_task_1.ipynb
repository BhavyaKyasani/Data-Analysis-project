```
frequency_df = df.groupby('index')['Order ID'].nunique().reset_index()
frequency_df.rename(columns={'Order ID': 'Frequency'}, inplace=True)
```

```
monetary_df = df.groupby('index')['TotalRevenue'].sum().reset_index()
monetary_df.rename(columns={'TotalRevenue': 'Monetary'}, inplace=True)

```

```
rfm = recency_df.merge(frequency_df, on='index')
rfm = rfm.merge(monetary_df, on='index')
```

```
rfm['R_Score'] = pd.qcut(rfm['Recency'].rank(method='first'), 5, labels=[5,4,3,2,1]).astype(int)
```

```
num_unique_frequencies = rfm['Frequency'].nunique()
if num_unique_frequencies < 5:
    # If there are fewer than 5 unique frequency values, qcut with 5 bins will fail.
    # Assign a default score (e.g., 1 for lowest frequency) for all entries.
    rfm['F_Score'] = 1
else:
    rfm['F_Score'] = pd.qcut(rfm['Frequency'].rank(method="first"), 5, labels=[1,2,3,4,5]).astype(int)
```

```
rfm['M_Score'] = pd.qcut(rfm['Monetary'].rank(method='first'), 5, labels=[1,2,3,4,5]).astype(int)
```

```
def rfm_segment(row):
    if row['R_Score'] >= 4 and row['F_Score'] >= 4 and row['M_Score'] >= 4:
        return 'High-Value Customer'
    elif row['R_Score'] >= 3 and row['F_Score'] >= 3:
        return 'Loyal Customer'
    elif row['R_Score'] <= 2 and row['F_Score'] >= 3:
        return 'At-Risk'
    elif row['R_Score'] <= 2 and row['F_Score'] <= 2:
        return 'Lost Customer'
    else:
        return 'Potential Customer'

rfm['Segment'] = rfm.apply(rfm_segment, axis=1)
```

```
rfm['Segment'] = rfm.apply(rfm_segment, axis=1)
segment_counts = rfm['Segment'].value_counts()
segment_counts
```

```
def get_segment(row):
    R, F, M = row['R_Score'], row['F_Score'], row['M_Score']

    # Adapting segmentation logic given that F_Score is 1 for all entries.
    # The original conditions for F_Score (e.g., F >= 3 or F >= 4) will not be met.
    # We redefine segments primarily using R and M scores, acknowledging F=1.

    if R >= 4 and M >= 4:
        return 'High-Value (R&M)'
    elif R >= 3 and M >= 3:
        return 'Potential Loyalists (R&M)'
    elif R == 5 and F == 1 and M >=3: # New & High Monetary
        return 'New & High Monetary'
    elif R <= 2 and M >= 3:
        return 'At-Risk (R&M)'
    elif R <= 2 and M <= 2:
        return 'Lost Customers (R&M)'
    else:
        return 'Other (R&M)'

rfm['Segment'] = rfm.apply(get_segment, axis=1)
```

```
top_categories = category_summary.head(5).index
plt.figure(figsize=(12,6))
for cat in top_categories:
    subset = category_growth[category_growth['Category']==cat]
    plt.plot(subset['Date'], subset['TotalRevenue'], label=cat)
plt.title("Sales Trends of Top 5 Product Categories")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

```
region_sales = df.groupby(['ship-state', pd.Grouper(key='Date', freq='M')])['TotalRevenue'].sum().reset_index()
region_summary = region_sales.groupby('ship-state')['TotalRevenue'].sum().sort_values(ascending=False)
print("Top Regions by Sales:")
print(region_summary.head(10))

# Plot regional trends
top_regions = region_summary.head(5).index
plt.figure(figsize=(12,6))
for reg in top_regions:
    subset = region_sales[region_sales['ship-state']==reg]
    plt.plot(subset['Date'], subset['TotalRevenue'], label=reg)
plt.title("Sales Trends in Top 5 Regions")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

```
snapshot_date = df['Date'].max() + pd.Timedelta(days=1)
rfm = df.groupby('index').agg(
    Date=('Date', lambda x: (snapshot_date - x.max()).days),  # Recency
    Frequency=('Order ID', 'count'),                            # Frequency (orders per index)
    Monetary=('TotalRevenue', 'sum')                                # Monetary
).rename(columns={'Date':'Recency'})
```

```
top_customers = rfm.sort_values(['Monetary', 'Frequency'], ascending=False).head(10)
print("Top 10 Customers for Upselling Opportunities:")
print(top_customers)
```

```
from itertools import combinations
from collections import Counter

pairs = []
for order_id, group in df.groupby('index'):
    products = group['Category'].unique() # Corrected from 'Product_Category'
    pairs.extend(combinations(products, 2))

pair_counts = Counter(pairs)
print("Top 10 Product Pairs for Cross-Selling:")
print(pair_counts.most_common(10))
```
